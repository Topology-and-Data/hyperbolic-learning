# Poincaré Embeddings
Exploring hyperbolic geometry and Poincaré embeddings, as per Nickel and Kiela from FAIR: [Poincaré  Embeddings for Learning Hierarchical Representations (2017)](https://arxiv.org/pdf/1705.08039.pdf).

Built off of the excellent [Python notebooks by Drew Wilimitis](https://github.com/drewwilimitis/hyperbolic-learning/blob/master/poincar%C3%A9_embeddings/poincare_notebook.ipynb).

```julia
using LinearAlgebra
using Distributions
using Plots
pyplot();
```

```julia
function poincare_dist(u, v, eps = 1e-5)
    d = 1 + 2 * norm.(u .- v) .^ 2 / ((1 .- norm.(u) .^ 2) * (1 .- norm.(v) .^2) + eps)
    return acosh(d)
end
```

```julia
function random_sample_circle(n)
    θ = rand(Uniform(0, 2π), n)
    u = rand(Uniform(0, 1), n)
    x = sqrt.(u) .* cos.(θ)
    y = sqrt.(u) .* sin.(θ)
    return x, y
end
```

```julia
random_sample = random_sample_circle(1000)
plot_1 = plot(
    random_sample, seriestype = :scatter, markersize = 3,
    aspect_ratio = 1, markercolor = :orange,
    markerstrokecolor = :white,
)

plot!(
    [0.0], [0.0], markersize = 4, seriestype = :scatter, markercolor = :gray25,
    legend = false,
)
```
